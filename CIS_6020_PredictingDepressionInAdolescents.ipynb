### Checking if any features can be dropped

1) Correlation heatmap - numerical variables

2) Chi-square test - categorical variables

```python
# Select numeric columns only, including 'At_Risk'
numeric_df = df_3.select_dtypes(include='number')

# Plot correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

# Manually select categorical columns (dtype = 'category')
categorical_cols = ['Gender', 'Parental_Control', 'Phone_Usage_Purpose']

# Chi-Square test for each pair
print("Chi-Square Test for Categorical Variable Associations:\n")
for i in range(len(categorical_cols)):
    for j in range(i + 1, len(categorical_cols)):
        var1 = categorical_cols[i]
        var2 = categorical_cols[j]
        table = pd.crosstab(df_3[var1], df_3[var2])

        if table.shape[0] > 1 and table.shape[1] > 1:  # Ensure valid table
            chi2, p, dof, _ = chi2_contingency(table)
            print(f"{var1:<25} <--> {var2:<25} | Chi2: {chi2:.2f} | p-value: {p:.4f}")

```

Correlation Heatmap (Numeric Features):
A correlation analysis was performed on all numeric features. The heatmap showed very weak correlations (all values close to 0), indicating that there is no strong linear relationship or redundancy among numeric variables. Therefore, no numeric features need to be removed due to multicollinearity.

Chi-Square Test (Categorical Features):
Chi-Square tests were conducted between all pairs of categorical variables. All p-values were greater than 0.05, suggesting no statistically significant association between categorical features. Hence, no categorical variables need to be dropped due to high similarity.

# Creating the Binary Depression Target Variable

The Target was created by using Depression_Level, where a person was consider at risk of depression if the score for the features were above 7.

1)Creating a binary target - If Depression > 7, then 'At Risk' (1) else (0)

2) Checking the distribution of the new target variable.

3) Dropping the original Depression and Anxiety Level columns.

```python
# Create binary anxiety target
df_4 = df_3.copy()

# 1 = At risk (Anxiety > 7), else 0
df_4['Depression_Target'] = np.where(df_3['Depression_Level'] > 7, 1, 0)
```

```python
# Anxiety target class distribution
print(df_4['Depression_Target'].value_counts())
print("\nDistribution (%):")
print(df_4['Depression_Target'].value_counts(normalize=True) * 100)
```

```python
df_4 = df_4.drop(['Depression_Level', 'Anxiety_Level'], axis=1)
```

## Handling the class-imbalance in the Target Variable

1) One-hot encoding categorical variables

2) Using oversampling technique - SMOTE to handle the class imbalance.

3) Splitting data in train and test datasets


```python
# Import required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Copy df_4 to df_6 and define X and y
df_6 = df_4.copy()
X = df_6.drop(columns=['Depression_Target', 'Depression_Level', 'Anxiety_Level'], errors='ignore')
y = df_6['Depression_Target']
```

```python
# One-hot encode categorical variables if any
X = pd.get_dummies(X, drop_first=True)
```

```python
# Split into training and test sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
```

```python
# Check shapes
print("Train before SMOTE:", X_train.shape, y_train.shape)
print("Train after SMOTE:", X_train_res.shape, y_train_res.shape)
print("Test shape:", X_test.shape, y_test.shape)
```

```python
# After SMOTE (resampled train)
print("\nTraining set class distribution (after SMOTE):")
print(y_train_res.value_counts(normalize=True))

# Test set
print("\nTest set class distribution:")
print(y_test.value_counts(normalize=True))
```

## ASE

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, mean_squared_error, classification_report, confusion_matrix
import matplotlib.pyplot as plt

# Step 1: Train the Decision Tree on Resampled Training Data
ase_tree = DecisionTreeClassifier(criterion='entropy', random_state=42, max_depth=4)
ase_tree.fit(X_train_res, y_train_res)

# Step 2: Plot the Tree
plt.figure(figsize=(18, 8))
plot_tree(
    ase_tree,
    feature_names=X.columns,
    class_names=['Not At Risk', 'Risk of Anxiety'],
    filled=True,
    rounded=True,
    fontsize=8
)
plt.title("ASE Decision Tree (Classification Tree with Entropy)")
plt.tight_layout()
plt.show()

# Step 3: Predict on Original Test Set
y_pred = ase_tree.predict(X_test)
y_proba = ase_tree.predict_proba(X_test)[:, 1]

# Step 4: Compute ASE and Accuracy
ase = mean_squared_error(y_test, y_proba)
accuracy = accuracy_score(y_test, y_pred)

print(" ASE Tree Evaluation Statistics:")
print(f" Accuracy: {accuracy:.4f}")
print(f" Average Squared Error (ASE): {ase:.4f}")

# Step 5: Classification Metrics
print("\n Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\n Classification Report:")
print(classification_report(y_test, y_pred))
```

## Misclass Tree

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, mean_squared_error, classification_report, confusion_matrix
import matplotlib.pyplot as plt

# Step 1: Train the Decision Tree using Gini Index (after SMOTE)
mis_tree = DecisionTreeClassifier(criterion='gini', max_depth=4, random_state=42)
mis_tree.fit(X_train_res, y_train_res)

# Step 2: Plot the Tree
plt.figure(figsize=(18, 8))
plot_tree(
    mis_tree,
    feature_names=X.columns,
    class_names=['Not At Risk', 'At Risk'],
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Misclassification Tree (Gini Index, max_depth=4)")
plt.tight_layout()
plt.show()

# Step 3: Evaluate the Tree
y_pred = mis_tree.predict(X_test)
y_proba = mis_tree.predict_proba(X_test)[:, 1]

ase = mean_squared_error(y_test, y_proba)
accuracy = accuracy_score(y_test, y_pred)

print(f"\nTree Evaluation Statistics (Misclassification Tree):")
print(f"- Accuracy: {accuracy:.4f}")
print(f"- Average Squared Error (ASE): {ase:.4f}")

# Step 4: Confusion Matrix and Classification Report
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, mean_squared_error, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Train Random Forest on resampled training set
rf_model = RandomForestClassifier(
    n_estimators=500,        # Number of trees
    max_depth=8,             # To reduce overfitting (adjustable)
    random_state=42,
    n_jobs=-1                # Use all CPU cores
)
rf_model.fit(X_train_res, y_train_res)

# Step 2: Predict on original test set
y_pred_rf = rf_model.predict(X_test)
y_proba_rf = rf_model.predict_proba(X_test)[:, 1]

# Step 3: Evaluate the model
ase_rf = mean_squared_error(y_test, y_proba_rf)
acc_rf = accuracy_score(y_test, y_pred_rf)

print("\nRandom Forest Evaluation:")
print(f"- Accuracy: {acc_rf:.4f}")
print(f"- Average Squared Error (ASE): {ase_rf:.4f}")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [4, 6, 8],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 3],
    'class_weight': ['balanced']  # Ensures weighting for class imbalance
}

# Initialize base model
rf_model = RandomForestClassifier(random_state=42)

# Setup GridSearchCV with recall as scoring
grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    scoring='recall',   # Focusing on recall
    cv=3,
    n_jobs=-1,
    verbose=2
)

# Fit to SMOTE-resampled training data
grid_search.fit(X_train_res, y_train_res)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, mean_squared_error, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Train Gradient Boosting on resampled training set
gb_model = GradientBoostingClassifier(
    n_estimators=300,
    learning_rate=0.1,
    max_depth=4,
    random_state=42
)
gb_model.fit(X_train_res, y_train_res)

# Step 2: Predict on original test set
y_pred_gb = gb_model.predict(X_test)
y_proba_gb = gb_model.predict_proba(X_test)[:, 1]

# Step 3: Evaluate model
ase_gb = mean_squared_error(y_test, y_proba_gb)
acc_gb = accuracy_score(y_test, y_pred_gb)

print("\nGradient Boosting Evaluation:")
print(f"- Accuracy: {acc_gb:.4f}")
print(f"- Average Squared Error (ASE): {ase_gb:.4f}")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_gb))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_gb))
```

```python
import os
import joblib
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, mean_squared_error, confusion_matrix, classification_report

# Path to save/load the best model
model_path = 'best_gb_model.pkl'

# Check if saved model exists
if os.path.exists(model_path):
    print("Loaded saved model from disk.")
    grid_search = joblib.load(model_path)
else:
    print("Running GridSearchCV on SMOTE data and saving the model...")

    # Define the parameter grid
    param_grid = {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 4, 5],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 3]
    }

    # Initialize the base model
    gb_model = GradientBoostingClassifier(random_state=42)

    # Initialize GridSearchCV
    grid_search = GridSearchCV(
        estimator=gb_model,
        param_grid=param_grid,
        scoring='recall',
        cv=3,
        n_jobs=-1,
        verbose=2
    )

    # Fit on SMOTE-resampled training data
    grid_search.fit(X_train_res, y_train_res)

    # Save best model
    joblib.dump(grid_search, model_path)
    print("Model saved to disk as:", model_path)

# Use the best estimator from grid search
best_gb_model = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)

# Predict on original training and test data
y_train_pred = best_gb_model.predict(X_train_res)  # Trained on resampled data
y_test_pred = best_gb_model.predict(X_test)
y_test_proba = best_gb_model.predict_proba(X_test)[:, 1]

# Evaluate performance
train_acc = accuracy_score(y_train_res, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)
ase = mean_squared_error(y_test, y_test_proba)

# Print evaluation
print("\n Tuned Gradient Boosting Evaluation:")
print(f"Training Accuracy: {train_acc:.4f}")
print(f"Test Accuracy:     {test_acc:.4f}")
print(f"Average Squared Error (ASE): {ase:.4f}")

# Reports
print("\n Classification Report (Train):")
print(classification_report(y_test, y_test_pred))
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# 1. Define parameter grid for XGBClassifier
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'subsample': [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9]
}

# 2. Initialize XGBClassifier
xgb_model = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    random_state=42,
    use_label_encoder=False # Suppress warning
)

# 3. Initialize StratifiedKFold for cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 4. Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    cv=cv,
    scoring='roc_auc',
    n_jobs=-1, # Use all available cores
    verbose=1
)

# 5. Fit GridSearchCV to the SMOTE-resampled training data
grid_search.fit(X_train_res, y_train_res)

# 6. Print the best parameters and best AUC score
print("Best Parameters:", grid_search.best_params_)
print("Best AUC Score:", grid_search.best_score_)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score

# Use the best estimator from grid search
best_xgb_model = grid_search.best_estimator_

# Predict on the original test set
y_pred_tuned_xgb = best_xgb_model.predict(X_test)
y_proba_tuned_xgb = best_xgb_model.predict_proba(X_test)[:, 1]

# Evaluate the model
print("\nTuned XGBoost Evaluation:")
print(f"- Accuracy: {accuracy_score(y_test, y_pred_tuned_xgb):.4f}")
print(f"- AUC Score: {roc_auc_score(y_test, y_proba_tuned_xgb):.4f}")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_tuned_xgb))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_tuned_xgb))
```

## Forward Stepwise Regression

```python
import statsmodels.api as sm
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score

# Step 0: Convert training data to DataFrame
X_df = pd.DataFrame(X_train_res_log, columns=X.columns)
y_df = y_train_res_log.reset_index(drop=True)

# Step 1: Define forward stepwise regression
def forward_stepwise(X, y, threshold=0.05):
    initial_features = []
    remaining_features = list(X.columns)
    selected_features = []

    while remaining_features:
        best_pval = float('inf')
        best_feature = None

        for feature in remaining_features:
            try:
                model = sm.Logit(y, sm.add_constant(X[initial_features + [feature]])).fit(disp=False)
                pval = model.pvalues[feature]
                if pval < best_pval:
                    best_pval = pval
                    best_feature = feature
            except:
                continue

        if best_feature and best_pval < threshold:
            initial_features.append(best_feature)
            remaining_features.remove(best_feature)
            print(f"Adding: {best_feature} (p-value = {best_pval:.4f})")
        else:
            break

    return initial_features

# Step 2: Run forward stepwise regression
selected_forward_features = forward_stepwise(X_df, y_df)

# Step 3: Fit final model on selected features
X_selected_train = sm.add_constant(X_df[selected_forward_features])
forward_model = sm.Logit(y_df, X_selected_train).fit()
print(forward_model.summary())

# Step 4: Prepare test set with same selected features
X_test_df_log = pd.DataFrame(X_test_scaled_log, columns=X.columns)
X_test_selected = X_test_df_log[selected_forward_features]
X_test_selected_sm = sm.add_constant(X_test_selected)

# Step 5: Predict and evaluate
y_test_pred_log = forward_model.predict(X_test_selected_sm)
y_test_pred_labels = (y_test_pred_log >= 0.5).astype(int)

print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_labels))
print("\nClassification Report:\n", classification_report(y_test, y_test_pred_labels))
print("Accuracy:", accuracy_score(y_test, y_test_pred_labels))
print("AUC Score:", roc_auc_score(y_test, y_test_pred_log))
```

## Backward Stepwise Regression

```python
import statsmodels.api as sm
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Step 0: Convert to DataFrame (if needed) to retain column names
X_df = pd.DataFrame(X_train_res_log, columns=X.columns)
y_df = y_train_res_log.reset_index(drop=True)  # ensure matching index

# Step 1: Add constant for intercept
X_log = sm.add_constant(X_df)
y_log = y_df.copy()

# Step 2: Backward elimination
def backward_stepwise(X, y, threshold=0.05):
    features = list(X.columns)
    while True:
        model = sm.Logit(y, X[features]).fit(disp=False)
        p_values = model.pvalues.drop('const', errors='ignore')

        if p_values.empty or p_values.max() <= threshold:
            break

        worst_feature = p_values.idxmax()
        print(f"Dropping: {worst_feature} (p-value = {p_values.max():.4f})")
        features.remove(worst_feature)
    return model, features

# Step 3: Run the backward stepwise regression
final_model, selected_features_from_backward = backward_stepwise(X_log, y_log)

# View final model summary
print(final_model.summary())

# Prepare test set with same selected features
X_test_df_log = pd.DataFrame(X_test_scaled_log, columns=X.columns)

# Ensure 'const' is added only once if needed and features match
X_test_selected_for_final = sm.add_constant(X_test_df_log[selected_features_from_backward[1:]]) # Exclude 'const' from selected_features_from_backward

# Step 1: Predict using final model
y_test_pred_log = final_model.predict(X_test_selected_for_final)
y_test_pred_labels = (y_test_pred_log >= 0.5).astype(int)

# Ensure index alignment for y_test as it's a Series
y_test_aligned = y_test.reset_index(drop=True)

# Step 2: Evaluate
print("Confusion Matrix:\n", confusion_matrix(y_test_aligned, y_test_pred_labels))
print("\nClassification Report:\n", classification_report(y_test_aligned, y_test_pred_labels))
print("Accuracy:", accuracy_score(y_test_aligned, y_test_pred_labels))
print("AUC Score:", roc_auc_score(y_test_aligned, y_test_pred_log))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

from pytorch_tabular import TabularModel
from pytorch_tabular.models.node.config import NodeConfig
from pytorch_tabular.config import DataConfig, TrainerConfig, OptimizerConfig, ExperimentConfig

# 1. Load the dataset
df = pd.read_csv('teen_phone_addiction_dataset.csv')

# Handle 'Addiction_Level_Binned_10' potentially not existing yet
# Create it temporarily if it doesn't exist, otherwise ensure it's dropped
if 'Addiction_Level_Binned_10' not in df.columns:
    df['Addiction_Level_Binned_10'] = pd.cut(df['Addiction_Level'], bins=10)

# 2. Create df_processed by dropping specified columns
df_processed = df.drop(columns=[
    'ID', 'Name', 'Location', 'School_Grade', 'Addiction_Level', 'Addiction_Level_Binned_10'
])

# 3. Convert specified columns to 'category' data type
df_processed['Gender'] = df_processed['Gender'].astype('category')
df_processed['Phone_Usage_Purpose'] = df_processed['Phone_Usage_Purpose'].astype('category')
df_processed['Parental_Control'] = df_processed['Parental_Control'].astype('category')

# 4. Create the 'Depression_Target' column
df_processed['Depression_Target'] = np.where(df_processed['Depression_Level'] > 7, 1, 0)

# 5. Drop original 'Depression_Level' and 'Anxiety_Level' columns
df_processed = df_processed.drop(columns=['Depression_Level', 'Anxiety_Level'])

# 6. Separate features (X) and target (y)
X = df_processed.drop(columns=['Depression_Target'])
y = df_processed['Depression_Target']

# 7. Apply one-hot encoding to X
X = pd.get_dummies(X, drop_first=True)

# 8. Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=42, stratify=y
)

# 9. Apply SMOTE to the training data
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# 10. Reconstruct DataFrames for pytorch_tabular
train_df_pfn = pd.concat([X_train_res, y_train_res.rename('Depression_Target')], axis=1)
test_df_pfn = pd.concat([X_test, y_test.rename('Depression_Target')], axis=1)

# Ensure the target column is float32 for metrics that expect probabilities
train_df_pfn['Depression_Target'] = train_df_pfn['Depression_Target'].astype(np.float32)
test_df_pfn['Depression_Target'] = test_df_pfn['Depression_Target'].astype(np.float32)

# 11. Identify continuous columns
continuous_cols = list(X_train_res.columns)

print("Data preprocessing complete. Variables prepared for pytorch_tabular:")
print(f"X_train_res shape: {X_train_res.shape}")
print(f"y_train_res shape: {y_train_res.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
print(f"train_df_pfn shape: {train_df_pfn.shape}")
print(f"test_df_pfn shape: {test_df_pfn.shape}")
print(f"Number of continuous columns: {len(continuous_cols)}")

# 1. Define DataConfig
data_config = DataConfig(
    target=["Depression_Target"],
    continuous_cols=continuous_cols,
)

# 2. Define NodeConfig
model_config = NodeConfig( # Correctly instantiate NodeConfig
    task="classification",
    head=None, # Set head to None as suggested by the warning
    metrics=["accuracy", "auroc"], # Removed 'f1'
    metrics_params=[{}, {"task": "multiclass", "num_classes": 2}], # Corrected auroc to expect multiclass output
    metrics_prob_input=[False, True]
)

# 3. Define TrainerConfig
trainer_config = TrainerConfig(
    accelerator="cpu",
    devices=1,
    max_epochs=10 # Reduced epochs for faster execution as per instruction
    # Removed 'callbacks=[]' as it is not an argument of TrainerConfig
)

# 4. Define OptimizerConfig (using default for now)
optimizer_config = OptimizerConfig(optimizer="Adam")

# 5. Define ExperimentConfig (optional)
experiment_config = ExperimentConfig(
    project_name="Depression_Risk_Prediction",
    run_name="pytorch_tabular_node_model",
    log_target="tensorboard",
)

# 6. Instantiate TabularModel
tab_model = TabularModel(
    data_config=data_config,
    model_config=model_config,
    optimizer_config=optimizer_config,
    trainer_config=trainer_config,
    experiment_config=experiment_config
)

# 7. Train the model
tab_model.fit(train=train_df_pfn, validation=test_df_pfn)

print("Tabular PFN model initialized and trained successfully.")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score

# Step 8: Predict on the test set
predictions = tab_model.predict(test_df_pfn)

# Extract predicted classes and probabilities
y_proba_pfn = predictions['Depression_Target_1.0_probability'].values # Probability of the positive class (class 1)
y_pred_pfn = (y_proba_pfn >= 0.5).astype(int) # Convert probabilities to binary class predictions

# Step 9: Evaluate the model
print("\nTabular PFN Evaluation:")
print(f"- Accuracy: {accuracy_score(y_test, y_pred_pfn):.4f}")
print(f"- AUC Score: {roc_auc_score(y_test, y_proba_pfn):.4f}")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_pfn))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_pfn))
```
